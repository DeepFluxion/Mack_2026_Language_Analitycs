## 1. Entendendo o Método `groupby` no Pandas

O método `groupby` é uma das funcionalidades mais importantes do pandas. Assim como as **tabelas dinâmicas no Excel**, o `groupby` permite que você:

- Agrupe seus dados com base em **uma ou mais colunas**
- Aplique **funções de agregação** — como soma, média e contagem — a cada grupo
- Obtenha resultados resumidos e organizados para análise

O processo segue sempre a mesma lógica de três etapas:

| Etapa | O que acontece |
|---|---|
| **Split** (Divisão) | Os dados são divididos em grupos com base no critério escolhido |
| **Apply** (Aplicação) | Uma função é aplicada a cada grupo independentemente |
| **Combine** (Combinação) | Os resultados de cada grupo são reunidos em uma estrutura final |

---
## 2. Como Funciona o `groupby`

Imagine que você tem um conjunto de dados de **vendas** e quer saber o total vendido por cada vendedor. Com o `groupby`, isso é simples:

```python
# Cria um DataFrame com dados de vendas por vendedor
data = {
    'Vendedor': ['Ana', 'Bruno', 'Ana', 'Carlos', 'Ana', 'Bruno'],
    'Vendas':   [200,   150,    340,   120,      300,   250]
}

df = pd.DataFrame(data)
df
```

```python
# Agrupa os dados pelo vendedor e soma as vendas de cada grupo
total_vendas = df.groupby('Vendedor').sum()
print(total_vendas)
```

Neste exemplo:
- O `DataFrame` `df` é **agrupado** pela coluna `'Vendedor'`
- A função `sum()` é **aplicada** para somar as vendas de cada vendedor
- O resultado é um novo `DataFrame` com o total de vendas por vendedor

> **Atenção:** perceba que a coluna `'Vendedor'` se torna o **índice** do DataFrame resultante.

---
## 3. Aplicando Diferentes Funções de Agregação

O `groupby` não se limita a somar valores. Você pode aplicar diversas funções de agregação para analisar seus dados:

| Função | O que faz |
|---|---|
| `sum()` | Soma os valores do grupo |
| `count()` | Conta o número de ocorrências |
| `mean()` | Calcula a média dos valores |
| `min()` | Encontra o valor mínimo |
| `max()` | Encontra o valor máximo |
| `std()` | Calcula o desvio padrão |

---
## 4. Exemplos Práticos

### 4.1 Contando o Número de Vendas por Vendedor

```python
# Conta quantas vendas cada vendedor realizou
contagem_vendas = df.groupby('Vendedor').count()
print(contagem_vendas)
```

### 4.2 Calculando a Média de Vendas por Vendedor

```python
# Calcula o valor médio das vendas por vendedor
media_vendas = df.groupby('Vendedor').mean()
print(media_vendas)
```

### 4.3 Encontrando a Venda Máxima por Vendedor

```python
# Encontra o maior valor de venda registrado por cada vendedor
venda_maxima = df.groupby('Vendedor').max()
print(venda_maxima)
```

### 4.4 Desvio Padrão e Valor Mínimo

```python
# Calcula o desvio padrão das vendas por vendedor
desvio_vendas = df.groupby('Vendedor').std()
print(desvio_vendas)
```

```python
# Encontra o menor valor de venda por vendedor
venda_minima = df.groupby('Vendedor').min()
print(venda_minima)
```

## 5. Dicas Importantes

### Dica 1 — Colunas não numéricas: use `numeric_only=True`

Se o seu `DataFrame` possuir colunas com valores não numéricos, utilize o argumento `numeric_only=True` ao chamar o método de agregação. Assim, você evita aplicar a agregação em colunas categóricas, o que pode levar a resultados indesejados ou erros.

```python
# DataFrame com colunas mistas: numérica e categórica
data_misto = {
    'Vendedor':  ['Ana', 'Bruno', 'Ana', 'Carlos', 'Ana', 'Bruno'],
    'Produto':   ['A',   'B',     'B',   'A',       'C',   'A'],
    'Vendas':    [200,   150,     340,   120,        300,   250]
}
df_misto = pd.DataFrame(data_misto)
df_misto
```

```python
# numeric_only=True: agrega apenas as colunas numéricas, ignora as categóricas
df_misto.groupby('Vendedor').sum(numeric_only=True)
```

### Dica 2 — O resultado é um `DataFrameGroupBy`

O resultado do `groupby` antes de aplicar uma função de agregação é um objeto especial do tipo `DataFrameGroupBy`. Você pode convertê-lo de volta para um `DataFrame` comum usando métodos como `reset_index()`:

```python
# O objeto groupby em si (sem agregação) é um DataFrameGroupBy
grupo = df.groupby('Vendedor')
print(type(grupo))
```

```python
# Após a agregação, o Vendedor vira índice do resultado
resultado = df.groupby('Vendedor').sum()
print("Com índice Vendedor:")
print(resultado)
print()

# reset_index() promove o índice de volta para coluna comum
print("Após reset_index():")
print(resultado.reset_index())
```

### Dica 3 — Agrupamento por múltiplas colunas

Você pode agrupar por **múltiplas colunas** passando uma lista de nomes de colunas para o `groupby`. O resultado terá um `MultiIndex`:

```python
# Agrupa por Vendedor e Produto ao mesmo tempo
grupo_multiplo = df_misto.groupby(['Vendedor', 'Produto']).sum(numeric_only=True)
print(grupo_multiplo)
```

```python
# reset_index() transforma o MultiIndex em colunas regulares
grupo_multiplo.reset_index()
```
