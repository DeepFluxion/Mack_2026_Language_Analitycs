O pandas fornece dois tipos de classes para lidar com dados:

## Seleção de dados

> **Nota:** Para código em produção, recomendamos usar os métodos otimizados de acesso a dados do pandas: `DataFrame.at()`, `DataFrame.iat()`, `DataFrame.loc()` e `DataFrame.iloc()`. Consulte a documentação de indexação: [Indexing and Selecting Data](https://pandas.pydata.org/docs/user_guide/indexing.html) e [MultiIndex / Advanced Indexing](https://pandas.pydata.org/docs/user_guide/advanced.html).

### Getitem (`[]`)

Para um `DataFrame`, passar um único rótulo seleciona uma coluna e retorna uma `Series`:

```python
# Seleciona a coluna 'A' como Series
df["A"]
```

Se o rótulo contiver apenas letras, números e underscores, é possível acessar a coluna como atributo:

```python
# Acesso à coluna como atributo (equivalente a df['A'])
df.A
```

Passando uma lista de rótulos de colunas, seleciona múltiplas colunas (útil para subconjuntos ou reordenação):

```python
# Seleciona múltiplas colunas passando uma lista
df[["B", "A"]]
```

Para um `DataFrame`, passar um slice `:` seleciona as linhas correspondentes:

```python
# Seleciona as 3 primeiras linhas por posição (0, 1, 2)
df[0:3]
```

```python
# Seleciona linhas por rótulo de data (ambos os extremos são incluídos)
df["20130102":"20130104"]
```

### Seleção por rótulo (`.loc`)

Consulte mais em [Selection by Label](https://pandas.pydata.org/docs/user_guide/indexing.html#selection-by-label) usando `DataFrame.loc()` ou `DataFrame.at()`.

Selecionando uma linha que corresponde a um rótulo:

```python
# Seleciona a primeira linha pelo seu rótulo de data
df.loc[dates[0]]
```

Selecionando todas as linhas (`:`) com colunas específicas:

```python
# Seleciona todas as linhas, apenas colunas 'A' e 'B'
df.loc[:, ["A", "B"]]
```

Fatiamento por rótulos — ambos os extremos são incluídos:

```python
# Seleciona intervalo de datas e colunas específicas (extremos incluídos)
df.loc["20130102":"20130104", ["A", "B"]]
```

Selecionando uma única linha e coluna retorna um escalar:

```python
# Seleciona valor escalar por rótulo de linha e coluna
df.loc[dates[0], "A"]
```

Para acesso rápido a um escalar (equivalente ao método anterior), use `df.at`:

```python
# Acesso rápido a escalar por rótulo (equivalente ao .loc acima)
df.at[dates[0], "A"]
```

### Seleção por posição (`.iloc`)

Consulte mais em [Selection by Position](https://pandas.pydata.org/docs/user_guide/indexing.html#selection-by-position) usando `DataFrame.iloc()` ou `DataFrame.iat()`.

Selecionando pela posição numérica:

```python
# Seleciona a 4ª linha (índice 3) por posição inteira
df.iloc[3]
```

```python
# Fatia linhas 3 a 4 e colunas 0 a 1 (semelhante ao NumPy/Python)
df.iloc[3:5, 0:2]
```

```python
# Seleção por listas de posições inteiras
df.iloc[[1, 2, 4], [0, 2]]
```

```python
# Fatiamento explícito de linhas (todas as colunas)
df.iloc[1:3, :]
```

```python
# Fatiamento explícito de colunas (todas as linhas)
df.iloc[:, 1:3]
```

```python
# Obtém valor escalar por posição de linha e coluna
df.iloc[1, 1]
```

```python
# Acesso rápido a escalar por posição (equivalente ao .iloc acima)
df.iat[1, 1]
```

### Indexação booleana

Selecionando linhas onde `df.A` é maior que 0:

```python
# Filtra apenas as linhas onde a coluna 'A' é positiva
df[df["A"] > 0]
```

Selecionando valores do `DataFrame` onde uma condição booleana é satisfeita (valores que não satisfazem se tornam `NaN`):

```python
# Mantém apenas valores positivos; valores negativos viram NaN
df[df > 0]
```

Usando o método `isin()` para filtrar:

```python
# Cria uma cópia e adiciona coluna categórica 'E'
df2 = df.copy()
df2["E"] = ["one", "one", "two", "three", "four", "three"]

# Filtra linhas cujo valor em 'E' está na lista fornecida
df2[df2["E"].isin(["two", "four"])]
```

## Atribuição (Setting)

Adicionar uma nova coluna alinha os dados automaticamente pelos índices:

```python
# Cria uma Series com índice deslocado e adiciona como coluna 'F'
s1 = pd.Series(
    [1, 2, 3, 4, 5, 6],
    index=pd.date_range("20130102", periods=6)
)
df["F"] = s1  # NaN onde o índice não coincide (2013-01-01)
```

```python
# Atribui valor por rótulo (linha e coluna)
df.at[dates[0], "A"] = 0

# Atribui valor por posição inteira
df.iat[0, 1] = 0

# Atribui um array NumPy inteiro à coluna 'D'
df.loc[:, "D"] = np.array([5] * len(df))

# Exibe o DataFrame atualizado
df
```

Uma operação `where` com atribuição — neste exemplo, todos os valores positivos são negados:

```python
df2 = df.copy()

# Onde os valores são positivos, substitui pelo valor negado
df2[df2 > 0] = -df2
df2
```

## Dados ausentes (Missing Data)

Para tipos NumPy, `np.nan` representa dados ausentes. Por padrão, ele não é incluído nos cálculos. Consulte a seção [Missing Data](https://pandas.pydata.org/docs/user_guide/missing_data.html).

A reindexação permite alterar/adicionar/excluir o índice em um eixo especificado. Retorna uma cópia dos dados:

```python
# Reindexação: reduz para 4 datas e adiciona a coluna 'E'
df1 = df.reindex(index=dates[0:4], columns=list(df.columns) + ["E"])

# Atribui 1 às duas primeiras linhas da coluna 'E'
df1.loc[dates[0]: dates[1], "E"] = 1
df1
```

`DataFrame.dropna()` remove qualquer linha que contenha dados ausentes:

```python
# Remove linhas que contêm pelo menos um NaN
df1.dropna(how="any")
```

`DataFrame.fillna()` preenche os dados ausentes com um valor:

```python
# Substitui NaN pelo valor 5
df1.fillna(value=5)
```

`isna()` retorna a máscara booleana onde os valores são `NaN`:

```python
# Retorna True onde o valor é NaN, False caso contrário
pd.isna(df1)
```

## Merge (Combinação de dados)

### Concat

O pandas oferece diversas facilidades para combinar `Series` e `DataFrames` com vários tipos de lógica de conjuntos para os índices. Consulte a seção [Merging](https://pandas.pydata.org/docs/user_guide/merging.html).

Concatenando objetos pandas por linhas com `concat()`:

```python
# Cria DataFrame de 10 linhas e 4 colunas com valores aleatórios
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# Divide o DataFrame em 3 pedaços
pieces = [df[:3], df[3:7], df[7:]]

# Concatena os pedaços de volta em um único DataFrame
pd.concat(pieces)
```

> **Nota:** Adicionar uma coluna a um `DataFrame` é relativamente rápido. Porém, adicionar uma linha requer uma cópia e pode ser custoso. Recomenda-se passar uma lista pré-construída de registros ao construtor do `DataFrame` em vez de construí-lo iterativamente.

### Join

`merge()` permite joins no estilo SQL em colunas específicas. Consulte a seção [Database style joining](https://pandas.pydata.org/docs/user_guide/merging.html#database-style-dataframe-or-named-series-joining-merging).

```python
# Merge com chaves duplicadas (produto cartesiano)
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})

pd.merge(left, right, on="key")
```

```python
# Merge com chaves únicas (1 para 1)
left = pd.DataFrame({"key": ["foo", "bar"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "bar"], "rval": [4, 5]})

pd.merge(left, right, on="key")
```

## Agrupamento (Grouping)

Por "group by" entendemos um processo que envolve uma ou mais das seguintes etapas:

1. **Dividir** os dados em grupos com base em algum critério.
2. **Aplicar** uma função a cada grupo independentemente.
3. **Combinar** os resultados em uma estrutura de dados.

Consulte a seção [Grouping](https://pandas.pydata.org/docs/user_guide/groupby.html).

```python
# Cria DataFrame com colunas categóricas e numéricas
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df
```

```python
# Agrupa pela coluna 'A', seleciona 'C' e 'D', e calcula a soma de cada grupo
df.groupby("A")[["C", "D"]].sum()
```

Agrupar por múltiplas colunas forma um `MultiIndex`:

```python
# Agrupa pelas colunas 'A' e 'B' e soma — resulta em MultiIndex
df.groupby(["A", "B"]).sum()
```

## Reshaping (Remodelagem)

Consulte as seções [Hierarchical Indexing](https://pandas.pydata.org/docs/user_guide/advanced.html) e [Reshaping](https://pandas.pydata.org/docs/user_guide/reshaping.html).

### Stack

O método `stack()` "comprime" um nível nas colunas do `DataFrame`:

```python
# Cria arrays para MultiIndex
arrays = [
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ["one", "two", "one", "two", "one", "two", "one", "two"],
]

# Cria o MultiIndex com nomes
index = pd.MultiIndex.from_arrays(arrays, names=["first", "second"])

# Cria DataFrame com MultiIndex
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=["A", "B"])

# Usa apenas as 4 primeiras linhas
df2 = df[:4]
df2
```

```python
# stack(): comprime as colunas para o índice (formato longo)
stacked = df2.stack()
stacked
```

Com um `DataFrame` ou `Series` "empilhado" (que possui `MultiIndex` como índice), a operação inversa de `stack()` é `unstack()`, que por padrão desempilha o último nível:

```python
# unstack(): operação inversa do stack — desempilha o último nível (padrão)
stacked.unstack()
```

```python
# Desempilha o nível 1 (segundo nível do índice)
stacked.unstack(1)
```

```python
# Desempilha o nível 0 (primeiro nível do índice)
stacked.unstack(0)
```

## Séries temporais (Time Series)

O pandas tem funcionalidades simples, poderosas e eficientes para realizar operações de reamostragem durante a conversão de frequência (por exemplo, converter dados por segundo em dados a cada 5 minutos). Isso é muito comum em aplicações financeiras. Consulte a seção [Time Series](https://pandas.pydata.org/docs/user_guide/timeseries.html).

```python
# Cria intervalo de 100 segundos e Series com inteiros aleatórios
rng = pd.date_range("1/1/2012", periods=100, freq="s")
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)

# Reamostra para intervalos de 5 minutos e soma os valores
ts.resample("5Min").sum()
```

`Series.tz_localize()` localiza uma série temporal em um fuso horário:

```python
# Cria série com 5 dias a partir de 06/03/2012
rng = pd.date_range("3/6/2012 00:00", periods=5, freq="D")
ts = pd.Series(np.random.randn(len(rng)), rng)

# Localiza no fuso horário UTC
ts_utc = ts.tz_localize("UTC")
ts_utc
```

`Series.tz_convert()` converte uma série temporal com fuso horário para outro fuso horário:

```python
# Converte de UTC para o fuso horário do leste dos EUA
ts_utc.tz_convert("US/Eastern")
```

Adicionando uma duração não fixa (`BusinessDay`) a uma série temporal:

```python
# Adiciona 5 dias úteis (pula fins de semana)
rng + pd.offsets.BusinessDay(5)
```

## Categorias (Categoricals)

O pandas pode incluir dados categóricos em um `DataFrame`. Para documentação completa, consulte a [introdução a dados categóricos](https://pandas.pydata.org/docs/user_guide/categorical.html) e a [documentação da API](https://pandas.pydata.org/docs/reference/arrays.html#categorical).

Convertendo notas brutas para o tipo de dado categórico:

```python
# Cria DataFrame com ID e nota bruta
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)

# Converte a coluna 'raw_grade' para tipo categórico
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

Renomeando as categorias para nomes mais significativos:

```python
# Renomeia as categorias existentes para nomes mais descritivos
new_categories = ["very good", "good", "very bad"]
df["grade"] = df["grade"].cat.rename_categories(new_categories)
```

Reordenando as categorias e simultaneamente adicionando as categorias faltantes:

```python
# Define a ordem completa das categorias (adiciona 'bad' e 'medium' que faltavam)
df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"]
)
df["grade"]
```

A ordenação respeita a ordem das categorias, não a ordem lexicográfica:

```python
# Ordena pelo valor da categoria (ordem definida, não alfabética)
df.sort_values(by="grade")
```

Agrupando por uma coluna categórica com `observed=False`, categorias vazias também são exibidas:

```python
# Com observed=False, mostra todas as categorias (incluindo as com contagem 0)
df.groupby("grade", observed=False).size()
```

## Plotagem (Plotting)

Consulte a [documentação de Plotting](https://pandas.pydata.org/docs/user_guide/visualization.html).

Usamos a convenção padrão para referenciar a API do matplotlib:

> **Nota:** No Jupyter, o gráfico aparecerá inline ao usar `plot()`. Em outros ambientes, use `matplotlib.pyplot.show()` para exibir ou `matplotlib.pyplot.savefig()` para salvar em arquivo.

```python
import matplotlib.pyplot as plt

# Fecha qualquer figura aberta anteriormente
plt.close("all")
```

```python
# Cria série com 1000 valores aleatórios e faz soma cumulativa
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()

# Plota a série temporal
ts.plot();
```

`plot()` plota todas as colunas:

```python
# Cria DataFrame de 1000 linhas e 4 colunas com soma cumulativa
df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["A", "B", "C", "D"]
)
df = df.cumsum()

# Cria nova figura e plota todas as colunas
plt.figure()
df.plot()

# Posiciona a legenda no melhor lugar disponível
plt.legend(loc='best');
```
