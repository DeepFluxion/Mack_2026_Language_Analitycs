## 1. Listas (`list`): O Trem de Dados
Como vimos na teoria, a lista é uma **sequência ordenada**. Imagine vagões de um trem onde cada posição (índice) guarda uma informação.

- **Quando usar:** Quando a ordem dos eventos importa (ex: logs cronológicos).
- **Característica:** Permite duplicatas e acesso pela posição.

```python
# Criando uma lista de sensores ativos
sensores = ["TEMP_01", "PRESS_02", "HUMID_03", "TEMP_01"] # Note a duplicata

# Acessando pelo índice (começa em 0)
print(f"Primeiro sensor instalado: {sensores[0]}")

# Adicionando um item ao final do 'trem'
sensores.append("LEVEL_04")
print(f"Lista atualizada: {sensores}")
```

## 2. Conjuntos (`set`): O Filtro de Unicidade
Diferente da lista, o Set é uma coleção de itens **únicos**. Pense nele como um filtro que remove automaticamente qualquer duplicata.

- **Quando usar:** Para deduplicação de bases ou verificações ultra-rápidas de existência.
- **Característica:** Não garante ordem, mas garante unicidade.

```python
# Queremos saber quais TIPOS de sensores temos, sem duplicatas
lista_suja = ["TEMP", "PRESS", "TEMP", "HUMID", "PRESS", "TEMP"]

# A "mágica" da conversão para Set
sensores_unicos = set(lista_suja)

print(f"Lista original (com duplicatas): {lista_suja}")
print(f"Set resultante (únicos): {sensores_unicos}")

# Verificação de existência (O Set é otimizado para isso)
print(f"Temos sensor de fluxo? {'FLOW' in sensores_unicos}")
```

## 3. Dicionários (`dict`): O Armário Etiquetado
Os dicionários são a base dos dados modernos (formato JSON). Eles guardam pares de **Chave: Valor**. Em vez de procurar por uma posição (como na lista), procuramos por uma "etiqueta" (chave).

- **Quando usar:** Para representar objetos complexos, perfis de usuários ou metadados de sistemas.

```python
# Representando um servidor da nossa infraestrutura
servidor_cloud = {
    "hostname": "aws-prod-data-01",
    "ip": "10.0.0.45",
    "status": "online",
    "tags": ["pipeline", "vendas", "v1"]
}

# Acessando uma informação específica pela etiqueta
print(f"O IP do servidor é: {servidor_cloud['ip']}")

# Adicionando uma nova informação
servidor_cloud["memoria_ram"] = "16GB"
print(f"Configuração completa: {servidor_cloud}")
```

## 4. O Impacto da Escolha: Performance e Big O
Como Engenheiros de Dados de um MBA, precisamos entender a **escalabilidade**.

Procurar um dado em uma **Lista** exige que o Python percorra item por item (Linear). Procurar em um **Set ou Dicionário** é instantâneo (O(1)), não importa se você tem 10 itens ou 10 milhões.

```python
import time

# Criando 1 milhão de IDs
base_dados_lista = list(range(1_000_000))
base_dados_set = set(base_dados_lista)

alvo = 999_999 # O último item (pior cenário)

# Testando a Lista
start = time.time()
alvo in base_dados_lista
print(f"Tempo na Lista: {(time.time() - start)*1000:.4f} ms")

# Testando o Set
start = time.time()
alvo in base_dados_set
print(f"Tempo no Set: {(time.time() - start)*1000:.4f} ms")
```

## 5. ESTUDO DE CASO: Limpeza de Logs JSON 🛠️
Na vida real, recebemos dados "sujos" em strings que precisam ser transformadas em dicionários para análise.

**O Problema:** Recebemos uma lista de strings simulando logs de geolocalização de um servidor. Precisamos converter essas strings em dicionários e filtrar apenas as que indicam erros (`ERROR`).

---
### O Engenheiro Aumentado (Uso de IA)
Não escreva o código do zero. Use o prompt abaixo no ChatGPT/Gemini para acelerar:

> "Atue como um Engenheiro de Dados. Tenho uma lista de strings onde cada string é um JSON: `logs = ['{\"level\": \"INFO\", \"msg\": \"ok\"}', '{\"level\": \"ERROR\", \"msg\": \"timeout\"}']`. Escreva uma função em Python que use `json.loads` para converter essas strings e retorne apenas uma lista com os dicionários cujo level seja 'ERROR'. Adicione comentários explicativos."

```python
import json

# Dados brutos recebidos do servidor
logs_brutos = [
    '{"level": "INFO", "timestamp": "2023-10-27T10:00:00Z", "msg": "Início da carga"}',
    '{"level": "ERROR", "timestamp": "2023-10-27T10:05:30Z", "msg": "Falha na conexão com o IP 192.168.1.5"}',
    '{"level": "DEBUG", "timestamp": "2023-10-27T10:10:00Z", "msg": "Limpando cache"}',
    '{"level": "ERROR", "timestamp": "2023-10-27T10:15:00Z", "msg": "Timeout no banco de dados"}'
]

# Implemente aqui a solução ou cole o código gerado pela IA:
def limpar_logs(lista_bruta):
    limpos = []
    for log_str in lista_bruta:
        dado = json.loads(log_str)
        if dado["level"] == "ERROR":
            limpos.append(dado)
    return limpos

# Resultado Final Estruturado
erros_detectados = limpar_logs(logs_brutos)
print(f"Detectamos {len(erros_detectados)} erros críticos:")
print(json.dumps(erros_detectados, indent=4))
```
